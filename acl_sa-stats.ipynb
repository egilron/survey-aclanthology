## Extract data from aclanthology


Ny pip package acl-anthology

```python
import sys
import numpy as np
import pandas as pd
from collections import Counter
import re, tqdm
from acl_anthology import Anthology
import warnings
warnings.filterwarnings('ignore')

anthology = Anthology.from_repo()
papers_count = len(list(anthology.papers()))
papers_count
```

```python
# FInd papers from certain year, with "sentiment" in title or abstract, check if semeval paper.
records = []
s_string = "sentiment" 
years = ["2020", "2022", "2024"] # include if any

for paper in tqdm.tqdm(anthology.papers(), total=papers_count):
    if (paper.year in years 
        and paper.pdf
        and any([s_string in str(paper.title).lower(),s_string in str(paper.abstract).lower() ])):
        
        records.append(dict(
            semeval = "semeval" in str(paper.title).lower(),
            id=paper.id,
            # paper=paper,
            bibkey=paper.bibkey,
            year=paper.year,
            url=paper.pdf.url
        ))
len(records)
```

```python
print("Semeval:",Counter(e["semeval"] for e in records))
display(records[0])
df = pd.DataFrame.from_records(records)
```

```python
df.to_parquet("sapapers.parquet")
```
